```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
stepsize = 0.1
num_revolutions = 5
```

```python
xs = np.arange(0, num_revolutions * 2 * np.pi, stepsize)
ys = np.sin(xs)
ys_1 = ys + 0.5
ys_2 = ys - 0.5
```

```python
print(len(xs))
```

```python
plt.figure(figsize=(15, 5))
plt.scatter(xs, ys, color='orange', s=5)
plt.fill_between(xs, ys_1, ys_2, color='grey', alpha=0.3)
plt.axis('off')
plt.show()
```
