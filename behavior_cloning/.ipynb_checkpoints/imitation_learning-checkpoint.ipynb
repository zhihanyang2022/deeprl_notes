```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader

import sys
sys.path.append('../../bishop1995_notes/modules_for_nn_training')
from callbacks import CallbackHandler
from learner import Learner
from trainer import Trainer
from utility_callbacks import LossCallback
```

## Problem 1: observations are non-markovian

```python
times = np.arange(0, 20, 0.5)
states = np.sin(times) * times
states /= np.max(np.abs(states))
```

```python
fig = plt.figure(figsize=(6, 4))
ax = fig.add_subplot(111, projection='3d')
ax.plot(times, states)
ax.view_init(70, 135)
ax.set_zticks([])
ax.set_title('Expert trajectory bird-eye view')
ax.set_xlabel('Observation axis 1')
ax.set_ylabel('Observation axis 2')
plt.tight_layout()
plt.show()
```

```python
x_train, y_train = states[:-1].reshape(-1, 1), states[1:].reshape(-1, 1)
print(x_train.shape, y_train.shape)
x_train, y_train = torch.from_numpy(x_train), torch.from_numpy(y_train).float()
train_ds = TensorDataset(x_train, y_train)
train_dl = DataLoader(train_ds, batch_size=39, shuffle=True)
```

```python
plt.scatter(x_train, y_train)
plt.show()
```

```python
class Regressor(nn.Module):
    
    def __init__(self, num_inputs, num_neurons):
        super().__init__()
        
        self.main = nn.Sequential(
            
            nn.Linear(num_inputs, num_neurons),
            nn.ReLU(),
            
            nn.Linear(num_neurons, num_neurons),
            nn.ReLU(),
            
            nn.Linear(num_neurons, num_neurons),
            nn.ReLU(),
            
            nn.Linear(num_neurons, num_neurons),
            nn.ReLU(),
            
            nn.Linear(num_neurons, num_neurons),
            nn.ReLU(),
            
            nn.Linear(num_neurons, 1)
        
        )
    
    def forward(self, xb):
        xb = self.main(xb)
        return xb.view(-1, 1)
```

```python
def get_model(num_inputs, num_neurons):
    nn = Regressor(num_inputs, num_neurons)
    return nn, optim.Adam(nn.parameters(), lr=1e-4)
```

```python
model, opt = get_model(1, 10)
loss = nn.MSELoss()
learn = Learner(train_dl, train_dl, model, loss, opt)

loss_cb = LossCallback()
cb_handler = CallbackHandler(cbs=[loss_cb])

trainer = Trainer(learn=learn, cb_handler=cb_handler)

trainer.find_lr(beta=0.98, final_value=10, num_itr=100)
```

```python
trainer.set_lr(1e-1)
```

```python
trainer.train(100)
```

```python
loss_cb.plot()
plt.show()
```

```python
model.eval()
plt.scatter(x_train, y_train)
plt.scatter(x_train, model(x_train.float()).detach().numpy())
plt.show()
```

```python
state = 0
states_gen = [state]
for i in range(len(times)):
    state = float(model(torch.Tensor([[state]]).float()).detach().numpy())
    states_gen.append(state)
```

```python
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot(times, states)
ax.plot(np.arange(0, 20+0.5, 0.5), states_gen)
ax.view_init(70, 135)
ax.set_zticks([])
ax.set_title('Expert trajectory bird-eye view', y=1.08)
ax.set_xlabel('Observation axis 1', labelpad=10)
ax.set_ylabel('Observation axis 2', labelpad=10)
plt.show()
```

```python
def rolling_window(a, window):
    shape = a.shape[:-1] + (a.shape[-1] - window + 1, window)
    strides = a.strides + (a.strides[-1],)
    return np.lib.stride_tricks.as_strided(a, shape=shape, strides=strides)
```

```python
window_size = 5
```

```python
x_train, y_train = rolling_window(states[:-1], window=window_size), states[np.arange(window_size, len(states))].reshape(-1, 1)
print(x_train.shape, y_train.shape)
x_train, y_train = torch.from_numpy(x_train), torch.from_numpy(y_train).float()
train_ds = TensorDataset(x_train, y_train)
train_dl = DataLoader(train_ds, batch_size=39, shuffle=True)
```

```python
model, opt = get_model(window_size, 100)
loss = nn.MSELoss()
learn = Learner(train_dl, train_dl, model, loss, opt)

loss_cb = LossCallback()
cb_handler = CallbackHandler(cbs=[loss_cb])

trainer = Trainer(learn=learn, cb_handler=cb_handler)

trainer.find_lr(beta=0.98, final_value=10, num_itr=100)
```

```python
trainer.set_lr(5e-3)
```

```python
trainer.train(100)
```

```python
loss_cb.plot()
```

```python
model.eval()
state = states[:window_size].copy().tolist()
states_gen = []
states_gen.extend(state)
for i in range(len(times) - window_size + 1):
    new_state = float(model(torch.Tensor([state]).float()).detach().numpy())
    state = state[1:] + [new_state]
    states_gen.append(new_state)
```

```python
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot(times, states)
ax.scatter(np.arange(0, 20+0.5, 0.5), states_gen, color='orange')
ax.view_init(70, 135)
ax.set_zticks([])
ax.set_title('Expert trajectory bird-eye view', y=1.08)
ax.set_xlabel('Observation axis 1', labelpad=10)
ax.set_ylabel('Observation axis 2', labelpad=10)
plt.show()
```
